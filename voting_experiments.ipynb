```python
from voting import *
```

```python
A = ['a', 'b', 'c', 'd', 'e']
E = [
    ('a', 'b', 3), 
    ('b', 'c', 3),
    ('c', 'a', 1),
    ('b', 'd', 5),
    ('d', 'a', 5),
    ('c', 'e', 5),
    ('e', 'b', 5),
    ]
G = MajorityGraph(nodes = A, edge_list = E)
G.draw()
```

```python
print(nx.find_cycle(G.G))
```

```python
V = VotingRules(G)
```

```python
V.compute_Slater_result()
V.show_Slater_result()
```

```python
V.compute_Kemeny_result()
V.show_Kemeny_result()
```

```python
from voting import *
```

```python
A = ['a', 'b', 'c', 'd', 'e', 'f']
E = [
    ('a', 'b'), 
    ('b', 'c'),
    ('c', 'd'),
    ('d', 'e'),
    ('e', 'f'),
    ('f', 'a'),
    ('e', 'b'),
    ('c', 'f')
    ]
G = MajorityGraph(nodes = A, edge_list = E)
G.draw()
```

```python
V = VotingRules(G)
```

```python
V.compute_Slater_result()
V.show_Slater_result()
```

```python
V.compute_Kemeny_result()
V.show_Kemeny_result()
```
